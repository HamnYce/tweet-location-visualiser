```python
stmt = 'SELECT lat,lng FROM public."Table_2018" WHERE hour = 1;'
with engine.connect() as conn:
    res = conn.execute(sqlalchemy.text(stmt))
    rows = res.all()
```

```python
def create_bar_fig():
    fig = px.scatter_mapbox(
        lat = [row['lat'] for row in rows],
        lon = [row['lng'] for row in rows],
        mapbox_style='open-street-map'
    )
    fig.show()

create_bar_fig()
```
